```python
import requests
import pandas as pd
from config import api_key
# Start with an empty list




Top_rated = []
# Loop through from pages 1 to 6
for i in range(1, 7):
    # Build the URL and download the results
    allurl = "https://api.themoviedb.org/3/movie/top_rated?api_key="+api_key+"&page=" + str(i)  
# apply request function
    req = requests.get(allurl)
    Top = req.json()
    results = Top['results']
    Top_rated.extend(results)
    #print(len(Top_rated))
    
df= pd.DataFrame(Top_rated)

df
```
